```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# I will read the Excel file first
df = pd.read_excel(r'C:\Users\ASUS\Downloads\merged_data2.xlsx')

# Converting all the columns to numeric type
df = df.apply(pd.to_numeric, errors='coerce')

# Filtering out non-numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Initializing the MinMaxScaler
scaler = MinMaxScaler()

# Normalizing the numeric columns
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])

# Plotting each numeric column individually
for column in numeric_columns:
    plt.figure(figsize=(10, 6))
    plt.plot(df[column])
    plt.xlabel('Data Points')
    plt.ylabel(column + ' (Normalized)')
    plt.title('Normalized Values of ' + column)
    plt.show()
```
