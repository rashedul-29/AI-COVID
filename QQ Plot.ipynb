```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Reading the Excel file
df = pd.read_excel(r'C:\Users\ASUS\Downloads\merged_data2.xlsx')

# Dropping the rows with missing values
df = df.dropna()

# Iterating over columns and creating all the QQ plots
for column in df.columns:
    # Handling all the non-numeric values
    df[column] = pd.to_numeric(df[column], errors='coerce')
    column_data = df[column]
    
    # Creating a QQ plot
    sm.qqplot(column_data, line='s')

    # Setting the plot title
    plt.title(f'QQ Plot of {column}')

    # Displaying the plot finally
    plt.show()
```
