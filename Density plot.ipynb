```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# For reading the Excel file
df = pd.read_excel(r'C:\Users\ASUS\Downloads\merged_data2.xlsx')

# Conversion of non-numeric values to NaN for all columns
df_numeric = df.apply(pd.to_numeric, errors='coerce')

# Now I will drop columns with non-numeric values
df_numeric = df_numeric.select_dtypes(include='number')

# I will iterate over each column now and create density plots one by one
for column in df_numeric.columns:
    # Selection of the column data
    data = df_numeric[column]
    
    # Removing all NaN values
    data = data.dropna()
    
    # Checking if the variance is zero or not
    if np.var(data) == 0:
        print(f"Skipped density plot for column: {column} (Zero variance)")
        continue
    
    # Creation of the density plot
    sns.kdeplot(data)
    
    # Setting all the plot labels and title
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.title(f'Density Plot of {column}')
    
    # Displaying the plot finally
    plt.show()
```
