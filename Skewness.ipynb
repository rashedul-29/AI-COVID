```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Reading the Excel file
df = pd.read_excel(r'C:\Users\ASUS\Downloads\merged_data2.xlsx')

# Converting all columns to numeric
df_numeric = df.apply(pd.to_numeric, errors='coerce')

# Calculating the skewness for numeric columns
skewness = df_numeric.apply(skew, axis=0, nan_policy='omit')

print("Skewness for each column:")
print(skewness)

# Visualizing all the data
for column in df_numeric.columns:
    if pd.api.types.is_numeric_dtype(df_numeric[column]):
        plt.hist(df_numeric[column].dropna(), bins=30, edgecolor='black')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.title(f'Distribution of {column}')
        plt.show()
```
